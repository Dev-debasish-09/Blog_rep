```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,3))
# Create a figure and axis
fig, ax = plt.subplots()

# Set axis titles
ax.set_title('Main Title')
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')

# Major and minor ticks
ax.tick_params(axis='both', which='major', length=10, color='blue')
ax.tick_params(axis='both', which='minor', length=5, color='red')

ax.legend(['Placeholder Legend'])

ax.grid(True)

ax.minorticks_on()

# Show the figure
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(5,3))
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
Xx,Yy = np.meshgrid(x,y)
plt.streamplot(x,y,Xx,Yy,color='blue',linewidth=1)
plt.suptitle('Streamplot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
fig.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# making data
x = np.random.randint(1,100,100)
y = np.random.randint(1,100,100)
z = np.random.randint(1,100,100)
#create a figure object
fig = plt.figure()
#create a axes object
ax = fig.add_subplot(111,projection='3d')
#plot scatter plot
ax.scatter(x,y,z,c=z,cmap='viridis',marker = 'o')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)
Xx,Yy = np.meshgrid(x,y)
zz = Xx**2 + Yy**2
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
p = ax.plot_surface(Xx,Yy,zz,cmap='viridis')
fig.colorbar(p)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
fig.show()
```
